```python

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python

# Modelo de regresión lineal multisalida
multi_output_model = LinearRegression()
multi_output_model.fit(df2018, df2019)

# Predicción y evaluación
y_multi_pred = multi_output_model.predict(df2018)
mse_por_planta = {}
r2_por_planta = {}

for i, col in enumerate(df2019.columns):
    mse_por_planta[col] = mean_squared_error(df2019[col], y_multi_pred[:, i])
    r2_por_planta[col] = r2_score(df2019[col], y_multi_pred[:, i])

# Resultados por planta
resultados_regresion_lineal = {
    "MSE por planta": mse_por_planta,
    "R² por planta": r2_por_planta
}

resultados_regresion_lineal
```

```python

# Realizar predicciones de la regresión lineal
y_pred = multi_output_model.predict(df2018)

# Crear subgráficas para cada planta
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
axs = axs.flatten()

for i, col in enumerate(df2019.columns):
    axs[i].scatter(df2019[col], y_pred[:, i], color='blue', alpha=0.7)
    axs[i].plot([df2019[col].min(), df2019[col].max()],
                [df2019[col].min(), df2019[col].max()],
                color='red', linestyle='--')
    axs[i].set_title(f"{col} - Real vs. Predicho")
    axs[i].set_xlabel("Real")
    axs[i].set_ylabel("Predicho")

plt.tight_layout()
plt.show()
```

Usando los datos de 2018, el modelo de regresión lineal predice muy bien los valores de 2019 para todas las plantas.
El ajuste es especialmente bueno para VIVEROS, LA CHINA y LA GAVIA (más del 84% de precisión).
LAS REJAS también se predice bien, aunque con un poco más de variación.

```python
from sklearn.ensemble import RandomForestRegressor

# Entrenar modelo Random Forest multisalida
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(df2018, df2019)

# Predecir y evaluar
y_rf_pred = rf_model.predict(df2018)
mse_rf = {}
r2_rf = {}

for i, col in enumerate(df2019.columns):
    mse_rf[col] = mean_squared_error(df2019[col], y_rf_pred[:, i])
    r2_rf[col] = r2_score(df2019[col], y_rf_pred[:, i])

# Resultados por planta con Random Forest
resultados_rf_multisalida = {
    "MSE por planta": mse_rf,
    "R² por planta": r2_rf
}

resultados_rf_multisalida
```

```python
# Realizar predicciones con el modelo Random Forest
y_rf_pred = rf_model.predict(df2018)

# Crear subgráficas para cada planta con Random Forest
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
axs = axs.flatten()

for i, col in enumerate(df2019.columns):
    axs[i].scatter(df2019[col], y_rf_pred[:, i], color='green', alpha=0.7)
    axs[i].plot([df2019[col].min(), df2019[col].max()],
                [df2019[col].min(), df2019[col].max()],
                color='red', linestyle='--')
    axs[i].set_title(f"{col} - Random Forest")
    axs[i].set_xlabel("Real")
    axs[i].set_ylabel("Predicho")

plt.tight_layout()
plt.show()
```

El modelo Random Forest multisalida tiene un desempeño excelente, aún mejor que la regresión lineal.
Predice todas las plantas con una precisión mayor al 93%, destacando LA GAVIA como la más fácil de anticipar.

```python
from sklearn.model_selection import train_test_split

# Separar en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(df2018, df2019, test_size=0.25, random_state=42)

# Entrenar el modelo con el conjunto de entrenamiento
rf_model_entrenado = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model_entrenado.fit(X_train, y_train)

# Predecir con el conjunto de prueba
y_test_pred = rf_model_entrenado.predict(X_test)

# Evaluar el modelo
mse_rf_test = {}
r2_rf_test = {}

for i, col in enumerate(df2019.columns):
    mse_rf_test[col] = mean_squared_error(y_test[col], y_test_pred[:, i])
    r2_rf_test[col] = r2_score(y_test[col], y_test_pred[:, i])

# Resultados del modelo entrenado y evaluado
resultados_entrenamiento_rf = {
    "MSE en prueba por planta": mse_rf_test,
    "R² en prueba por planta": r2_rf_test
}

resultados_entrenamiento_rf
```

Aunque Random Forest funcionó muy bien sin validación, su rendimiento baja cuando se entrena con solo el 75% de los datos.

LA GAVIA sigue siendo la más predecible.

Las demás plantas muestran sobreajuste, lo que es común con conjuntos de datos pequeños.

```python
# Crear un gráfico de barras comparando R² en prueba por planta
plt.figure(figsize=(8, 5))
plt.bar(r2_rf_test.keys(), r2_rf_test.values(), color='skyblue', edgecolor='black')
plt.axhline(y=0, color='red', linestyle='--', label='Línea base (R² = 0)')
plt.title("Desempeño del modelo Random Forest (R² en conjunto de prueba)")
plt.ylabel("R²")
plt.xlabel("Planta")
plt.legend()
plt.tight_layout()
plt.show()
```

Cuanto más alto esté cada barra (más cerca de 1), mejor fue la predicción.

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.multioutput import MultiOutputRegressor
import numpy as np

# Configurar validación cruzada con 5 particiones (K-Fold)
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Usar Random Forest dentro de un envoltorio multisalida
rf_multi_cv_model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))

# Calcular puntajes R² por planta (se promedian los resultados por cada salida)
r2_scores_cv = cross_val_score(rf_multi_cv_model, df2018, df2019, cv=cv, scoring='r2')

# Resultado medio y desviación estándar
resultado_cv = {
    "R² promedio": np.mean(r2_scores_cv),
    "R² desviación estándar": np.std(r2_scores_cv),
    "R² por partición": r2_scores_cv.tolist()
}

resultado_cv
```

```python
# Crear gráfico de barras para mostrar R² por partición de la validación cruzada
plt.figure(figsize=(8, 5))
plt.bar(range(1, 6), resultado_cv["R² por partición"], color='orange', edgecolor='black')
plt.axhline(y=0, color='red', linestyle='--', label='Línea base (R² = 0)')
plt.title("R² por partición - Validación cruzada con Random Forest")
plt.xlabel("Partición")
plt.ylabel("R²")
plt.xticks(range(1, 6))
plt.legend()
plt.tight_layout()
plt.show()
```

El rendimiento es muy inestable entre particiones.

El modelo funciona bien en algunos subconjuntos, pero se rompe completamente en otros (R² = –2.80).

Esto se debe a que el dataset es pequeño (solo 24 muestras) y no hay suficientes ejemplos variados para cada división.
